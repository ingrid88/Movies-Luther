```python
import re

def match_line(i,line):
    floating_point = "[0-9]*\.?[0-9]*"
    integer_dots = "[0-9\.\*]+"
    integer = "[0-9]+"
    
    try:
        match = re.match('\s*('+integer_dots+')\s+('+integer+')\s+('+floating_point+')\s*(.+)', line).groups()
        return match
    except:
        print str(i) + ":" + line
        raise 
```
