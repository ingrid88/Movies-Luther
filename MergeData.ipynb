```python
import pickle
import numpy as np
import pandas as pd
```

```python
with open("processedMojo.pkl", 'r') as picklefile: 
    mojo = pickle.load(picklefile)
    
with open("oscars.pkl", 'r') as picklefile: 
    oscars = pickle.load(picklefile)
    
with open("imdb.pkl", 'r') as picklefile: 
    imdb = pickle.load(picklefile)
    
with open("budget.pkl", 'r') as picklefile: 
    budget = pickle.load(picklefile)
```

```python
imdb
```

```python
imdb_and_mojo = pd.merge(mojo,imdb,on="Title")
```

```python
oscars.head()
```

```python
oscars_and_imdb_mojo = pd.merge(oscars,imdb_and_mojo,how='right',on='Title').drop({'Year_y','Studio_x'},axis=1)
oscars_and_imdb_mojo.Noms.fillna(value=0,inplace=True)
oscars_and_imdb_mojo.Wins.fillna(value=0,inplace=True)
```

```python
new_oscars_and_imdb_mojo = oscars_and_imdb_mojo.drop({"Opening_no_Theaters","Gross_no_Theaters","Year_x"},axis=1)
```

```python
new_oscars_and_imdb_mojo.head()
```

```python
budget_mojo_imdb = pd.merge(budget,new_oscars_and_imdb_mojo,right_on="Title",left_on="Movie").drop({"Unnamed: 0","YrRank","Worldwide Gross","Movie","Release Date","Writers"},axis=1)
```

```python
budget_mojo_imdb = budget_mojo_imdb[budget_mojo_imdb.Actors.apply(lambda actors: len(actors)) == 3]
```

```python
budget_mojo_imdb["Actor1"] = budget_mojo_imdb.Actors.apply(lambda actors: actors[0])
budget_mojo_imdb["Actor2"] = budget_mojo_imdb.Actors.apply(lambda actors: actors[1])
budget_mojo_imdb["Actor3"] = budget_mojo_imdb.Actors.apply(lambda actors: actors[2])
```

```python
budget_mojo_imdb.drop("Actors",axis=1)
```

```python
budget_mojo_imdb.rename(columns={'Production Budget': 'Production_Budget', 'Domestic Gross': 'Domestic_Gross', 'Studio_y':'Studio'}, inplace=True)
# budget_mojo_imdb.columns = ["Production_Budget","Domestic_Gross","cpi_date","Title","Box_Office","Noms","Wins","Rank","Studio","Total_Gross","Opening","Open","IMDB_Score","Users","Metascore","Director","Writers","Actors","Genres","Rating","Actor1","Actor2","Actor3"]
```
